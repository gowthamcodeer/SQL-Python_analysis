```python
#download dataset using kaggle api
import zipfile
zip_ref = zipfile.ZipFile('orders.csv.zip')
zip_ref.extractall()#extract file to dir
zip_ref.close() #close file
```

```python
#read the data from file and handling null values
import pandas as pd
df = pd.read_csv('orders.csv',na_values = ['Not Available','unknown'])
```

```python
df.head(1)
```

```python
df['Ship Mode'].unique()
```

```python
#rename columns names ..make them lower case and replace space with underscore
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
```

```python
df.head(1)
```

```python
#derive new columns discount.sales price and profit
df['discount'] = df['list_price'] * df['discount_percent']*0.01
df['sale_price'] = df['list_price'] - df['discount']
df['profit'] = df['sale_price']-df['cost_price']
```

```python
df.head()
```

```python
df.dtypes
```

```python
#convert order date from object data type to datetime
df['order_date'] = pd.to_datetime(df['order_date'],format = '%Y-%m-%d')
```

```python
df.dtypes
```

```python
df.head(1)
```

```python
#drop columns which not needed
df.drop(columns = ['list_price','cost_price','discount_percent'],inplace = True)
```

```python
df.head(1)
```

```python
#finding na and replacing it with most occuring value
df.isna().sum()
```

```python
df['ship_mode'].mode()
```

```python
df['ship_mode'].fillna(df['ship_mode'].mode()[0],inplace = True)
```

```python
from sqlalchemy import create_engine

# Database connection details
username = 'root'
password = '1234'
host = '127.0.0.1'  # or your MySQL server address
port = '3306'       # default MySQL port
database = 'orders'

# Create the connection string
connecting_string = f"mysql+mysqldb://{username}:{password}@{host}:{port}/{database}"
# or for mysqlclient
# connecting_string = f"mysql+mysqldb://{username}:{password}@{host}:{port}/{database}"

# Create the SQLAlchemy engine
engine = create_engine(connecting_string)
```

```python
conn = engine.connect()
```

```python
#load the data in mysql
df.to_sql('df_orders',con = conn,index = False,if_exists = 'append')
```
